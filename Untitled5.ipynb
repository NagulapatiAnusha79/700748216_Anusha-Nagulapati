```python
from keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load data
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Reshape to [samples][width][height][channels]
train_images_reshaped = train_images.reshape((train_images.shape[0], 28, 28, 1)).astype('float32')
test_images_reshaped = test_images.reshape((test_images.shape[0], 28, 28, 1)).astype('float32')

# Normalize from 0-255 to 0-1
train_images_normalized = train_images_reshaped / 255
test_images_normalized = test_images_reshaped / 255

# One hot encode outputs
train_labels_encoded = to_categorical(train_labels)
test_labels_encoded = to_categorical(test_labels)
num_classes = test_labels_encoded.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten

custom_model = Sequential()
custom_model.add(Conv2D(32, (5, 5), input_shape=(28, 28, 1), activation='relu'))
custom_model.add(MaxPooling2D(pool_size=(2, 2)))
custom_model.add(Flatten())
custom_model.add(Dense(128, activation='relu'))
custom_model.add(Dense(num_classes, activation='softmax'))

# Compile model
custom_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
from keras.callbacks import ReduceLROnPlateau

# Callbacks
lr_reduce = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.001)

# Train
training_history = custom_model.fit(train_images_normalized, train_labels_encoded, validation_data=(test_images_normalized, test_labels_encoded), epochs=10, batch_size=200, callbacks=[lr_reduce], verbose=2)
```

```python
evaluation_scores = custom_model.evaluate(test_images_normalized, test_labels_encoded, verbose=0)
print("CNN Error: %.2f%%" % (100 - evaluation_scores[1] * 100))
```

```python
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam

# Load and preprocess data
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
train_images = train_images.reshape((train_images.shape[0], 28, 28, 1)).astype('float32') / 255
test_images = test_images.reshape((test_images.shape[0], 28, 28, 1)).astype('float32') / 255
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

def create_model(learning_rate=0.001):
    model = Sequential([
        Conv2D(32, (5, 5), input_shape=(28, 28, 1), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax')
    ])
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Hyperparameters
learning_rates = [0.001, 0.0001]
batch_sizes = [32, 64]

# Grid search
best_accuracy = 0
best_lr = 0
best_batch_size = 0

for lr in learning_rates:
    for batch_size in batch_sizes:
        model = create_model(learning_rate=lr)
        history = model.fit(train_images, train_labels, validation_data=(test_images, test_labels), epochs=5, batch_size=batch_size, verbose=0)
        val_accuracy = np.max(history.history['val_accuracy'])
        print(f"Learning Rate={lr}, Batch Size={batch_size}, Validation Accuracy={val_accuracy}")

        if val_accuracy > best_accuracy:
            best_accuracy = val_accuracy
            best_lr = lr
            best_batch_size = batch_size

print(f"Best parameters: Learning Rate={best_lr}, Batch Size={best_batch_size}, with validation accuracy={best_accuracy}")
```

```python
import matplotlib.pyplot as plt

# Plotting Accuracy
plt.plot(training_history.history['accuracy'])
plt.plot(training_history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='upper left')
plt.show()

# Plotting Loss
plt.plot(training_history.history['loss'])
plt.plot(training_history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='upper left')
plt.show()
```
